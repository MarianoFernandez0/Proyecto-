```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets_dir = 'error_results/urbano'
files = sorted(os.listdir(datasets_dir))
frecuencies = ['7Hz', '15Hz', '30Hz', '40Hz', '50Hz', '60Hz']
algorithms = ['NN', 'GNN', 'PDAF', 'JPDAF', 'ENNJPDAF']

print(files)
print(frecuencies)
print(algorithms)
```

```python
datasets = []
for file in files:
    tracks = pd.read_csv(os.path.join(datasets_dir, file), index_col=0)
    datasets.append(tracks)
    print('Number of tests', file, tracks.shape[0])
tracks.head(-1)
```

Dataset 1:

```python
results = datasets[0]
results_att_sort = []
for att in algorithms: 
    idxs = [idx for idx in tracks.index if att == idx.split(sep='_')[1]]
    results_att = results.loc[idxs, :]
    results_att_sort.append(results_att)
results_att.head(-1)
```

```python
for i, results_att in enumerate(results_att_sort):
    for idx in results_att.index:
        frec = idx.split(sep='_')[0]
        frec = frec[:-2]
        results_att.loc[idx, 'Frecuency'] = int(frec) 
    
results_att.head()
```

OSPA:

```python
plt.figure(figsize=(20, 10))
for i, results_att in enumerate(results_att_sort):
    results_att = results_att.sort_values(by=['Frecuency'])
    plt.plot(results_att['Frecuency'], results_att['OSPA'], marker='o', markersize=5,
             label=results_att.index[0].split(sep='_')[1])
plt.legend()
plt.grid()
plt.title('OSPA - Frecuency:')
plt.show()
# results_att.plot(y='OSPA', x='Frecuency', subplots=True,figsize=(20,5),
# use_index=True, legend=True, rot=0, grid=True, title=('OSPA - Frecuency:'))
```

```python
plt.figure(figsize=(20, 10))
for i, results_att in enumerate(results_att_sort):
    results_att = results_att.sort_values(by=['Frecuency'])
    plt.plot(results_att['Frecuency'], results_att['JSC Tracks'], marker='o', markersize=5,
             label=results_att.index[0].split(sep='_')[1])
plt.legend()
plt.grid()
plt.title('JSC Tracks - Frecuency:')
plt.show()
```
