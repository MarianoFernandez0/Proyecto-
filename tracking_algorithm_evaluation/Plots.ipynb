```python
import pandas as pd
import numpy as np
```

```python
tracks = pd.read_csv('performance_measures_octave.csv', index_col=0)
tracks.head(10)
```

```python
dataset = 'dataset_1_10Hz'
bool_idx = []
for idx in tracks.index:
    bool_idx.append(dataset in idx)
tracks_dataset_1_10Hz = tracks.loc[bool_idx, :]
tracks_dataset_1_10Hz.head()
```

```python
tracks_dataset_1_10Hz.plot.bar(y='JSC Tracks', subplots=True,figsize=(20,5),
                               use_index=True, legend = False, rot=0)
```
