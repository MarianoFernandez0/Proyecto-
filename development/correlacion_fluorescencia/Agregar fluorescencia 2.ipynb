```python
import pandas as pd
import numpy as np
```

```python
entrada = pd.read_csv("in.csv")
salida = pd.read_csv("out.csv")
```

```python
# ver la cantidad de puntos detectados para un frame en el df de entrada y salida

for i in range (salida['frame'].max()+1):

    if len(salida.loc[salida['frame'] == i]) != 0: #que exista el frame
        
        # armo dataframes de entrada y salida que contienen solo un frame
        new_in = entrada.loc[entrada['frame'] == i] 
        new_out = salida.loc[salida['frame'] == i]
        
        if new_in.shape[0] > new_out.shape[0]:
            print(new_in.shape[0],new_out.shape[0])
```

```python
cost = np.zeros([salida.shape[0],entrada.shape[0]])
for it_out in range(salida.shape[0]):
    
    x_out, y_out = salida.loc[it_out, "x"], salida.loc[it_out, "y"]
    
    for it_in in range(entrada.shape[0]):

        x_in, y_in = entrada.loc[it_in, "x"], entrada.loc[it_in, "y"]
        distance = np.sqrt((x_out-x_in)**2 + (y_out-y_in)**2)
        cost[it_out,it_in] = distance
```

```python
from scipy.optimize import linear_sum_assignment
row_ind, col_ind = linear_sum_assignment(cost)
row_ind
```
