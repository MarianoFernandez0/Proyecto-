```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# load df
who_measures = pd.read_csv('input/Archive/measures/dataset_1/dataset_1_30Hz_ennjpdaf_WHO.csv')
who_measures.head()
```

```python
# remove id col
measures = who_measures.drop('track_id', axis=1)
measures.head()
```

```python
measures_array = measures.to_numpy()
for i in range(1, 5):
    plt.figure(figsize=(15, 10))
    y_pred = KMeans(n_clusters=i).fit_predict(measures)
    plt.scatter(measures_array[:, 0], measures_array[:, 1], c=y_pred, s=25)
    plt.title(f"Number of Clusters: {i}")
plt.show();
```
