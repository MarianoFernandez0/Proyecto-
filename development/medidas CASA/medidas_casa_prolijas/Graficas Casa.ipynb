```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
casa_c = np.loadtxt(fname='features_who_nn.csv', delimiter=',')
```

```python
input_data=casa_c

vcl = input_data[:,0]
vsl = input_data[:,1]
vap = input_data[:,2]
alh = input_data[:,3]
lin = input_data[:,4]
wob = input_data[:,5]
stra = input_data[:,6]
bcf = input_data[:,7]
mad = input_data[:,8]
```

```python
## Graficas
fig = plt.figure(figsize=(20,5))

plt.subplot(141),plt.plot(vsl, vcl, 'or' ),plt.xlabel("VSL"),plt.ylabel("VCL"),plt.grid(True, which="both")
plt.subplot(142),plt.plot(alh, lin, 'or' ),plt.xlabel("ALH"),plt.ylabel("LIN"),plt.grid(True, which="both")
plt.subplot(143),plt.plot(vsl, wob, 'or' ),plt.xlabel("VSL"),plt.ylabel("WOB"),plt.grid(True, which="both")
plt.subplot(144),plt.plot(mad, lin, 'or' ),plt.xlabel("MAD"),plt.ylabel("LIN"),plt.grid(True, which="both")
plt.savefig(fname='grafica_nn.png')
```
