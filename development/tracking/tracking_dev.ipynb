```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

'''
Voy a intentar hacer un tracking utilizando la JPDAF.
'''

#Cargo datos
df_detections = pd.read_csv("./input_data/salida_sigma_0_0_data.csv")
img = mpimg.imread("./input_data/salida_sigma_0_0.png")
```

```python
frames = int(df_detections["frame"].max())
tracks = []
coordinates_frames = df_detections[["x", "y", "frame"]]

#df['My new column'] = 'default value'
#for column in df:

#inicializo los tracks
founded = False
first_frame_with_detection = -1

while not founded and first_frame_with_detection < frames:
    first_frame_with_detection += 1
    coordinates_df = coordinates_frames[coordinates_frames["frame"] == first_frame_with_detection]
    coordinates = coordinates_df[["x", "y"]].to_numpy()
    if coordinates.shape[0] > 0:
        founded = True
        for coord in range(coordinates.shape[0]):
            #Las filas son frames y las columnas las coordenadas
            track_coord_list = np.empty((frames,2))
            track_coord_list[:] = None
            track_coord_list[first_frame_with_detection, :] = coordinates[coord, :]
            tracks.append(track_coord_list)
    
    
#for frame in range(first_frame_with_detection, frames):
    

        
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img, cmap = "gray")
for track in tracks:
    plt.plot(track[0,1], track[0,0], "or")
plt.show()
```
