```python
import imageio
import matplotlib.pyplot as plt
import skimage.color as color
import cv2
import numpy as np
import scipy
```

```python
video = imageio.mimread('1472 semen-00.avi')
```

```python
im_gray = cv2.cvtColor(video[0], cv2.COLOR_BGR2GRAY)
gaussian1 = cv2.GaussianBlur(im_gray,(11,11),0)
gaussian2 = cv2.GaussianBlur(gaussian1,(11,11),0)


fig = plt.figure(figsize=(17,7))
plt.subplot(131),plt.imshow(im_gray,cmap='gray')
plt.subplot(132),plt.imshow(gaussian1,cmap='gray')
plt.subplot(133),plt.imshow(gaussian2,cmap='gray')
```

```python
# FILTRO GAUSSIANO
it_gaussian_filter = 3
prev_img = im_gray
for it in range(it_gaussian_filter):
    gaussian = cv2.GaussianBlur(prev_img,(11,11),0)
    prev_img = gaussian
    
    
fig = plt.figure(figsize=(17,5))
plt.subplot(121),plt.imshow(im_gray,cmap='gray')
plt.subplot(122),plt.imshow(prev_img,cmap='gray')

cv2.imshow('a7',prev_img)
cv2.waitKey(0)
```

```python
# FILTRO LAPLACIANO
dst = cv2.Laplacian(prev_img, ddepth=cv2.CV_16S, ksize=9)
abs_dst = cv2.convertScaleAbs(dst)
cv2.imshow('a7',abs_dst)
cv2.waitKey(0)
```
